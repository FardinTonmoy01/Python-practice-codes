```python
class node:
  def __init__(self,node_no,prev_node,actual_cost,total_cost):
    self.node_no=node_no
    self.prev_node=prev_node
    self.actual_cost=actual_cost
    self.total_cost=total_cost

  def backtrack(self):
   
    current= self.node_no
    prev= self.prev_node
    print(current)
    print('<-')
    print(prev)
    print('<-')
    current=prev
    prev= current.prev_node
  def nodes(self):
    return self.node_no
    



s1=node(0,None,0,0+h_value[0])

a1=node(1,0,1,1+h_value[1])

b1=node(2,0,4,4+h_value[2])
b2=node(2,1,3,3+h_value[2])

c1=node(3,1,6,6+h_value[3])
c2=node(3,2,6,6+h_value[3])
c3=node(3,2,5,5+h_value[3])

d1=node(4,3,6,6+h_value[4])
d2=node(4,3,8,8+h_value[4])

```
