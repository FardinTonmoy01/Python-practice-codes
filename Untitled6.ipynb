```python
import pandas as pd
import csv
from sklearn.feature_extraction.text import CountVectorizer
from datetime import datetime

# Change these paths as needed
data_path = "/emails.csv"
output_path = "processed_enron_email_dataset.csv"

# Load data
# df = pd.read_csv(data_path, error_bad_lines=False)
df = pd.read_csv(data_path, quoting=csv.QUOTE_NONE, escapechar='\\')


# Remove emails with missing values
df.dropna(inplace=True)

# Convert text data to numerical features using Bag-of-words
vectorizer = CountVectorizer(max_features=1000)
features_bow = vectorizer.fit_transform(df["content"])

# Create features for timeliness
df["sent_date"] = pd.to_datetime(df["sent_date"])
time_diff = (datetime.now() - df["sent_date"]).dt.total_seconds()

# Create features for conciseness
word_count = df["content"].apply(lambda x: len(x.split()))

# Combine features
features = pd.concat([features_bow, time_diff, word_count], axis=1)

# Download processed data
features.to_csv(output_path, index=False)

print("Processed data saved to:", output_path)
```
