```python
!apt-get install -y libmagickwand-dev
!pip install Wand

from google.colab import files
from wand.image import Image

# Upload the EPS file from your local machine to Colab
uploaded = files.upload()

# After running the file uploader, use the file name to set the input EPS file path
input_eps_file = next(iter(uploaded))

# Output folder path where PNG files will be saved
output_folder = 'https://drive.google.com/drive/folders/1iMTx2Jw7QjlmRmJxPQ3B10IRnsYYFoBo?usp=drive_link'

# Read the uploaded EPS file and convert each page to a PNG
with Image(blob=uploaded[input_eps_file]) as img:
    # Iterate through each page in the EPS file
    for i, page in enumerate(img.sequence):
        # Set the resolution and read the page
        page.resolution = 300  # Set your desired resolution
        with Image(page) as img_page:
            img_page.format = 'png'
            # Save each page as a separate PNG file
            output_path = f"{output_folder}page_{i + 1}.png"
            img_page.save(filename=output_path)

# Download the generated PNG files
for i in range(len(img.sequence)):
    files.download(f"{output_folder}page_{i + 1}.png")
```
