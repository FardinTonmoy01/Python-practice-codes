```python

import numpy as np

from numpy import genfromtxt
data_path= '/content/drive/MyDrive/Ai lab/iris.csv'
data1= genfromtxt(data_path, delimiter=',')
```

```python
data= data1.tolist()
print(data)
```

```python
Train_set=[]
```

```python
val_set=[]
```

```python
Test_set=[]
```

```python
import random
random.shuffle(data)
s=[]
ss=[]
```

```python
s=data
ss=data
print(s)
print(ss)

classes=[]
classes_traning=[]
```

```python
#print(R)
for i in s:
    r= random.uniform(0,1)
    if r>=0 and r<=0.70:
     Train_set.append(i)
    elif r>0.70 and r<=0.85:
      val_set.append(i)
    else:
      Test_set.append(i)


print(Train_set) 
print(val_set) 
print(Test_set)

```

```python
for i in val_set:
  x= i[-1]
  classes.append(x)
  del i[-1]
print(val_set)
```

```python
for pp in Train_set:
  xx= pp[-1]
  classes_traning.append(xx)
  del pp[-1]
print(Train_set)
```

```python
#for k=7 knn classification
k=7
L=[]
i=1
from scipy.spatial import distance
for v in val_set:
  for t in Train_set:
    tx= distance.euclidean(v,t)
    L.append(tx)

print(L)
L.sort()
print(L)
first_k= L[0:k]
print(first_k)

from collections import Counter
m= Counter(first_k).most_common(1)
majority=m[0][0]
print(majority)

c=0 
for i in classes:
  if (i== majority):
    print('yes')
    break

for j in classes:
 if (j== majority):
    c=c+1

Val_acc=(c/len(val_set))*100
print(Val_acc)
```
